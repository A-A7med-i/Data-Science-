# Students performance in exams

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
pass_mark=40
```

##  Read the data and get information 

```python
df =pd.read_csv("StudentsPerformance.csv")
df.head()
```

```python
df.describe()
```

```python
size=df.size
shape=df.shape
print(f"The size is '{size}' and the rows is '{shape[0]}' and the columns is '{shape[1]}' ")
```

## Get the null values

```python
df.isnull().sum()
```

## Graph the math score 

```python
sns.countplot(data=df,x="math score",palette="tab10") 
plt.show()
```

## Student pass in math score

```python
df["Math_PassStatus"]=np.where(df["math score"]>pass_mark,"P","F")
df.Math_PassStatus.value_counts()
```

```python
graph=sns.countplot(data=df,x="parental level of education",hue="Math_PassStatus",palette="tab10")
_=plt.setp(graph.get_xticklabels(),rotation=90)
plt.show()
```

## Graph reading score

```python
colors="#AF7AC5"
sns.histplot(data=df,x="reading score",binwidth =1.3,color=colors)
plt.show()
```

## Student pass in reading exam

```python
df["Reading_PassStatus"]=np.where(df["reading score"]>pass_mark,"P","F")
df.Reading_PassStatus.value_counts()
```

```python
graph=sns.countplot(data=df,x="parental level of education",hue="Reading_PassStatus",palette="Dark2")
_=plt.setp(graph.get_xticklabels(),rotation=90)
plt.show()
```

## Graph writing score

```python
sns.countplot(data=df,x="writing score",palette ="Set1")
plt.show()
```

## Student pass in writing score

```python
df["Writing_PassStatus"]=np.where(df["writing score"]>pass_mark,"P","F")
df.Writing_PassStatus.value_counts()
```

```python
graph=sns.countplot(x='parental level of education', data = df, hue='Writing_PassStatus', palette='Set1')
_=plt.setp(graph.get_xticklabels(),rotation=90)
plt.show()
```

## Let's apply multiple plots

```python
graph=sns.FacetGrid(df,col="parental level of education",hue="race/ethnicity",col_wrap=3,height=5,palette="Set1")
graph.map(plt.scatter,"math score","race/ethnicity")
plt.show()
```

## Student pass in all sub. 

```python
df["OverAll_PassStatus"]=df.apply(lambda x: "P"if x["Math_PassStatus"]=="P"
                                               and x["Reading_PassStatus"]=="P"
                                               and x["Writing_PassStatus"]=="P" else "F",axis=1)
df.OverAll_PassStatus.value_counts()
```

## Graph the student pass in overall sub.

```python
graph=sns.countplot(data=df,x="parental level of education",hue="OverAll_PassStatus",palette="Set2")
_=plt.setp(graph.get_xticklabels(),rotation=90)
plt.show()
```

## Get the total marks and percentage marks  

```python
df["Total_Marks"]=df["math score"]+df["reading score"]+df["writing score"]
df["Percentage"]=df["Total_Marks"]/3
```

```python
sns.displot(df,x="Percentage",hue="race/ethnicity",kind="kde",fill=True,palette="Set1")
plt.show()
```

## Grading

```python
def Get_Grade(Percentage, OverAll_PassStatus):
    if OverAll_PassStatus =="P":
        if Percentage>=80:
            return "A"
        elif Percentage>=70 and Percentage<=80:
            return "B"
        elif Percentage>=60 and Percentage<=70:
            return "C"
        elif Percentage>=50 and Percentage<=60:
            return "D"
        elif Percentage>=40 and Percentage<=50:
            return "E"
    else:
        return "F"
df["Grades"]=df.apply(lambda x:Get_Grade(x["Percentage"],x["OverAll_PassStatus"]),axis=1)
df.Grades.value_counts().sort_index()
```

```python
sns.catplot(data=df,x="Grades",kind="count",order=['A','B','C','D','E','F'],palette="muted")
```

```python
graph=sns.countplot(data=df,x="parental level of education",hue="Grades",palette='summer')
_=plt.setp(graph.get_xticklabels(),rotation=90)
```
