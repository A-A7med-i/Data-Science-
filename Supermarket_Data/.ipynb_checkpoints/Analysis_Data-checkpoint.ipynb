# Introduction 
<a id="1"></a>

##### The growth of supermarkets in most populated cities are increasing and market competitions are also high.
##### The dataset is one of the historical sales of supermarket company which has recorded in 3 different branches for 3 months data.
##### Predictive data analytics methods are easy to apply with this dataset.

#### The following is important to understand the data
##### Attribute information
   - Invoice id: Computer generated sales slip invoice identification number
   - Branch: Branch of supercenter (3 branches are available identified by A, B and C).
   - City: Location of supercenters
   - Customer type: Type of customers, recorded by Members for customers using member card and Normal for without member card.
   - Gender: Gender type of customer
   - Product line: General item categorization groups - Electronic accessories, Fashion accessories, Food and beverages, Health    - and beauty, Home and lifestyle, Sports and travel
   - Unit price: Price of each product in $
   - Quantity: Number of products purchased by customer
   - Tax: 5% tax fee for customer buying
   - Total: Total price including tax
   - Date: Date of purchase (Record available from January 2019 to March 2019)
   - Time: Purchase time (10am to 9pm)
   - Payment: Payment used by customer for purchase (3 methods are available â€“ Cash, Credit card and Ewallet)
   - COGS: Cost of goods sold
   - Gross margin percentage: Gross margin percentage
   - Gross income: Gross income
   - Rating: Customer stratification rating on their overall shopping experience (On a scale of 1 to 10)


# Python Libraries
<a id="2"></a>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
%matplotlib inline 
sns.set_style("white")
sns.set_theme(style="whitegrid")
```

# Read Data And Get Information
<a id="3"></a>

```python
df=pd.read_csv("supermarket_sales.csv")
df.head()
```

```python
df.describe()
```

```python
shape=df.shape
size=df.size
print(f"the rows is: {shape[0]} and the columns is: {shape[1]}")
print(f"the size is: {size}")
```

# Cleaning Data
<a id="4"></a>

```python
df.isnull().sum()
```

### Above cells' conclusion
- no missing value
- we need to remove some coloumn 

```python
df["Date"]=pd.to_datetime(df.Date)
df["Date"]
```

```python
df["Month"]=df["Date"].dt.month_name()
df["Month"]
```

```python
df.drop(columns="Invoice ID",inplace=True)
```

```python
df.head()
```

### From Above 
- we drop the col Invoice ID
- we added col Month
- we convert the date to date time

# Visualization Data
<a id="5"></a>

```python
df.Branch.value_counts()
```

```python
graph=sns.countplot(data=df,x="Branch",order=df.Branch.value_counts().index,palette="Set2")
plt.show()
```

### From Above 
- 3 Branch Founded
- A branch is the largest
- B branch is the least

```python
df["Customer type"].value_counts()
```

```python
sns.countplot(data=df,x="Customer type",order=df["Customer type"].value_counts().index,palette="Set1")
plt.show()
```

### From Above 
- There is a strong convergence between member and normal

```python
df.Gender.value_counts()
```

```python
sns.countplot(data=df,x="Gender",order=df.Gender.value_counts().index,palette="tab20")
plt.show()
```

### From Above 
- There is a strong convergence between male and female

```python
df["Product line"].value_counts()
```

```python
graph=sns.countplot(data=df,x="Product line",order=df["Product line"].value_counts().index,palette="Set1")
_=plt.setp(graph.get_xticklabels(),rotation=90)
plt.show()
```

### From Above 
- Fashion accessories is the most product sold
